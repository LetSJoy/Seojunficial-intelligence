## Numpy 실습

- 대용량 계산에서 흔히 사용되는 Numpy는 C로 구현되어 파이썬의 특징인 dynamic typing을 포기하는 대신 성능을 확보했다.

- 일반적으로 처리 속도는 for loop < list comprehension < Numpy이다. 이를 증명하는 프로그램을 작성하시오.

```python
import numpy as np

iteration = 1000000 # 1,000,000
scalar = 2

# for loop
def loops(scalar, vector):
    result=[]
    for value in vector:
        result.append(value * scalar)
    return result

%timeit loops(scalar, list(range(iteration)))
%timeit result2 = [i*scalar for i in range(iteration)]
%timeit result3 = 2*np.array(range(iteration))
```

## (선택과제)

numpy는 수학관련 함수를 다양하게 제공하고 있다. numpy 문서(https://numpy.org/doc/1.18/reference/routines.math.html) 탐색을 통해 함수 3개를 선정하고 관련 예제를 작성하시오.

```python
# 1. prod():axis 방향의 원소들의 곱을 하나 낮은 dimention의 array로 return
a = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(np.prod(a,0))
print("===========")
print(np.prod(a,1))
print("===========")
print(np.prod(a,2))
```

```python
# 2. cumprod(): axis를 따라서 한칸씩 움직이며 지금까지의 누적 곱을 return 차원을 낮추지 않고 array의 값을 바꿈
a = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(np.cumprod(a,0))
print("===========")
print(np.cumprod(a,1))
print("===========")
print(np.cumprod(a,2))
```

```python
# 3. diff(): axis를 따라서 n번 반복하여 연속한 두 원소의 차이를 반환 -> axis방향으로 크기 한칸 감소
# 횟수의 변화를 준 경우우
a = np.array([[[1,2,3],[3,4,5]],[[5,6,7],[7,8,9]]])
print(np.diff(a,0))
print("===========")
print(np.diff(a,1))
print("===========")
print(np.diff(a,2))
print("===========\n===========")
# axis의 변화를 준 경우
b = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(np.diff(b,1,0))
print("===========")
print(np.diff(b,1,1))
print("===========")
print(np.diff(b,1,2))
```
